```python
# 일반경기 배팅 데이터 전처리_08_02_01
import pandas as pd
import os

# 폴더 내의 모든 엑셀 파일을 읽어올 경로
folder_path = 'data_good/'

# 모든 엑셀 파일을 저장할 빈 리스트 생성
all_data_frames = []

# 폴더 내의 모든 엑셀 파일을 반복해서 처리합니다.
for i, file_name in enumerate(os.listdir(folder_path)):
    if file_name.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file_name)

        # 엑셀 파일을 읽어옵니다.
        xls = pd.ExcelFile(file_path)

        # 각 시트를 DataFrame으로 읽어와서 리스트에 저장합니다.
        sheet_names = xls.sheet_names
        data_frames = [xls.parse(sheet_name) for sheet_name in sheet_names]

        # 모든 데이터를 하나의 DataFrame으로 통합합니다.
        merged_data = pd.concat(data_frames, ignore_index=True)
        # 'ModifyTime' 열의 데이터를 datetime 형식으로 변환합니다.
        merged_data['ModifyTime'] = pd.to_datetime(merged_data['ModifyTime'], format='%H:%M:%S', errors='coerce')
        # 'ModifyTime' 열을 기준으로 데이터를 시간 순서로 정렬합니다.
        merged_data = merged_data.sort_values(by='ModifyTime', ascending=True)
        # 모든 데이터프레임을 리스트에 추가합니다.
        all_data_frames.append(merged_data)

        # CSV 파일로 저장합니다.
        csv_file_name = os.path.join(folder_path, f"merged_good_data_{i+1:02}.csv")
        merged_data.to_csv(csv_file_name, index=False)
        print(csv_file_name, "을 통합 저장하였습니다.")
```

```python
# 배당률 통합 파일 차트 그리기_08_02_02
file_path = 'data_good/merged_good_data_07.csv'
merged_data = pd.read_csv(file_path)

# 차트 그리기
import pandas as pd
import matplotlib.pyplot as plt
# 'ModifyTime'을 x 축으로, 'HomeWin', 'Standoff', 'GuestWin'을 y 축으로 하는 라인 차트를 그립니다.
plt.figure(figsize=(12, 6))
plt.plot(merged_data['ModifyTime'], merged_data['HomeWin'], label='HomeWin', marker='o')
plt.plot(merged_data['ModifyTime'], merged_data['Standoff'], label='Standoff', marker='o')
plt.plot(merged_data['ModifyTime'], merged_data['GuestWin'], label='GuestWin', marker='o')
# 그래프 제목과 레이블을 추가합니다.
plt.title('HomeWin, Standoff, and GuestWin Over Time')
plt.xlabel('ModifyTime')
plt.ylabel('Odds')
plt.legend()
plt.autoscale(enable=True, axis='both')  # 'both'로 설정하여 x와 y 축을 모두 자동으로 조정합니다.
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 이동평균으로 그래프 스무싱_제공되지 않은 코드

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
file_path = 'data_good/merged_good_data_07.csv'
merged_data = pd.read_csv(file_path)

# 'ModifyTime'을 datetime 형식으로 변환
merged_data['ModifyTime'] = pd.to_datetime(merged_data['ModifyTime'])

# 데이터를 일정 기간에 대한 이동 평균으로 스무스하게 만듭니다.
window_size = 10  # 이동 평균의 기간을 설정합니다.
smoothed_home_win = merged_data['HomeWin'].rolling(window=window_size).mean()
smoothed_standoff = merged_data['Standoff'].rolling(window=window_size).mean()
smoothed_guest_win = merged_data['GuestWin'].rolling(window=window_size).mean()

# 차트 그리기
plt.figure(figsize=(12, 6))
plt.plot(merged_data['ModifyTime'], smoothed_home_win, label='HomeWin', marker='o')
plt.plot(merged_data['ModifyTime'], smoothed_standoff, label='Standoff', marker='o')
plt.plot(merged_data['ModifyTime'], smoothed_guest_win, label='GuestWin', marker='o')

# 그래프 제목과 레이블을 추가합니다.
plt.title('Smoothed HomeWin, Standoff, and GuestWin Over Time')
plt.xlabel('ModifyTime')
plt.ylabel('Smoothed Odds')
plt.legend()
plt.autoscale(enable=True, axis='both')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
